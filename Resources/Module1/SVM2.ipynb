What if the training data is linearly non-separable?
- Introduce slack variables $\xi\in \mathbb{R}^N$ to allow misclassification for difficult examples or outliers.

- relax constraints: $y^{(i)}( \mathbf{x}^{(i)}\mathbf{w}+w_0)\ge 1-\xi_i, \xi_i\ge 0$. (soft margin constraint)

-  If $\xi_i\ge 1$, $\mathbf{x}^{(i)}$ is misclassified. If $0<\xi_i< 1$, $\mathbf{x}^{(i)}$ is inside the margin, but on the correct side of the decision boundary.


<img src="https://github.com/yiboxu20/MachineLearning/blob/main/Resources/images/soft_margin_xi.png?raw=true" width="300" />


### Soft-margin classification
The optimization problem becomes
$$\boxed{\min_{\mathbf{w}\in\mathbb{R}^d, w_0\in \mathbb{R},\xi\in \mathbb{R}^N}\frac{1}{2}\|\mathbf{w}\|^2+ \color{red}{\lambda \sum_{i=1}^N \xi_i}} $$
$$\text{subject to } \mathbf{y}^{(i)}( \mathbf{x}^{(i)}\mathbf{w}+w_0)\ge 1 \color{red}{-\xi_i}, $$
$$ \color{red}{\xi_i\ge 0}, \forall i=1, \dots, N$$

$\lambda>0$ is pre-determined regularization parameter.
$\frac{\xi_i}{\|\mathbf{\mathbf{w}}\|}$ is the distance $\mathbf{x}^{(i)}$ goes across the corresponding margin boundary.  
- small $\lambda$ encourages large $\xi_i$: allows misclassification with large margin.

- large $\lambda$ encourages small $\xi_i$: discourages misclassification with narrow margin.

- $\lambda=\infty$ enforces $\xi_i=0$: hard margin allows no misclassification.


### Dual for soft-margin case

$$\boxed{\max_{\alpha\in \mathbb{R}^N} \sum_{i=1}^N \alpha_i-\frac{1}{2}\sum_{i=1}^N \sum_{j=1}^N y^{(i)}y^{(j)}\alpha_i\alpha_j \mathbf{x}^{(i)}(\mathbf{x}^{(j)})^\top} $$
$$\text{subject to } \sum_{i=1}^N \alpha_i \mathbf{y}^{(i)}=0, 0\le \alpha_i \color{red}{\le \lambda}, \forall i=1,\dots, N $$
Added upper bound of $\lambda$ on $\alpha_i$.

Compare with primal vs dual problem.

**Primal**:
- $d+N+1$ unknowns.
- $N$ linear inequality constraints, $N$ non-negativity constraints.
- convex quadratic programming.

**Dual**:
- $N$ unknowns.
- one linear equality constraint, $N$ box constraints.
- The convex quadratic programming to minimize the objective function
$\frac{1}{2}\alpha^\top \mathbf{G}\alpha -\mathbb{1}^\top \alpha$, where $\mathbf{G}_{ij}=y^{(i)}y^{(j)} \mathbf{x}^{(i)}(\mathbf{x}^{(j)})^\top $ and $\mathbf{G}$ is called Gram matrix.
$$ \mathbf{v}=\begin{bmatrix}y^{(1)}\mathbf{x}^{(1)}\\ \vdots \\ y^{(N)}\mathbf{x}^{(N)} \end{bmatrix}.$$
$$ \mathbf{G}=\mathbf{v} \mathbf{v}^\top.$$

## Revisit Perceptron problem

Training perceptron on dataset $\{\mathbf{x}^{(i)}, y^{(i)}\}_{i=1}^N\subset \mathbf{R}^d\times \{\pm 1\}$:
$$\boxed{\min_{\mathbf{w\in \mathbb{R}^d, w_0\in \mathbb{R}}}\sum_{i=1}^n\left(y^{(i)}-\phi(\mathbf{x}^{(i)}\mathbf{w} +w_0)\right)^2} $$
where the activation function $\phi(z)=1$ if $z>0$, $\phi(z)=-1$, otherwise.

In other words, we tried to find parameters such that the signs of the linear output $\mathbf{x}^{(i)}\mathbf{w} +w_0$ and $y^{(i)}$ agree. Equivalently,
 $$\boxed{\min_{\mathbf{w\in \mathbb{R}^d, w_0\in \mathbb{R}}}\sum_{i=1}^n \mathbb{1}_{\{y^{(i)}(\mathbf{x}^{(i)}\mathbf{w} +w_0)< 0\}} } $$
where the indicator function
$$ \mathbb{1}_{\{z<0\}}\begin{cases}1 & \text{if } z< 0 \\ 0 & \text{Otherwise}\end{cases}$$

This is also called **0-1 loss**.

This type of problem is intractable to minimize: discontinuous (discrete-valued), non-convex,
non-uniqueness of global min.


## SVM with hinge loss



Let's rephrase this again

$$\boxed{\min_{\mathbf{w}\in\mathbb{R}^d, w_0\in \mathbb{R},\xi\in \mathbb{R}^N}\frac{1}{2}\|\mathbf{w}\|^2+ \lambda \sum_{i=1}^N \xi_i }$$
$$\text{subject to } \xi_i\ge \max\left\{0, 1-\mathbf{y}^{(i)}(\mathbf{x}^{(i)}\mathbf{w} +w_0)\right\}, \forall i=1, \dots, N$$

- If $\xi_i=0>1-\mathbf{y}^{(i)}(\mathbf{x}^{(i)}\mathbf{w} +w_0)$, $\mathbf{x}^{(i)}$ is **correct** side of margin
boundary.

- If $\xi_i=1-\mathbf{y}^{(i)}( \mathbf{x}^{(i)}\mathbf{w} +w_0)>0$, $\mathbf{x}^{(i)}$ is **wrong** side of margin
boundary.


We can simply choose $\xi_i= \max\left\{0, 1-\mathbf{y}^{(i)}(\mathbf{x}^{(i)}\mathbf{w} +w_0)\right\}$ for $\forall i=1, \dots, N$.



### Alternative formulation

Drop the constraints and replace $\xi_i= \max\left\{0, 1-\mathbf{y}^{(i)}( \mathbf{x}^{(i)}\mathbf{w} +w_0)\right\}$.

$$ \boxed{\min_{\mathbf{w}\in \mathbb{R}^d, w_0\in \mathbb{R}} \underbrace{\lambda\sum_{i=1}^N \max\left\{0, 1-\mathbf{y}^{(i)}( \mathbf{x}^{(i)}\mathbf{w} +w_0)\right\} +\frac{1}{2} \|\mathbf{w}\|_2^2}_{=\text{LOSS}(\mathbf{w},w_0)}}$$

- $\ell(z)=\max\{1-z, 0\}$ is called **hinge loss**. It is a convex relaxation to 0-1 loss. Hinge loss depends on how far the examples are from the decision boundary.

- Hinge loss can be minimized via subgradient method

- Soft-margin SVM minimize the hinge loss with Tikhonov regularization ($l_2$ regularization).


<img src="https://github.com/yiboxu20/MachineLearning/blob/main/Resources/images/loss_comparison.png?raw=true" width="500" />

The gradient of the loss function is
$$\nabla_{\mathbf{w}} \text{LOSS}(\mathbf{w},w_0) =\mathbf{w}+\sum_{i=1}^N\begin{cases} 0 & \text{if } \mathbf{y}^{(i)}( \mathbf{x}^{(i)}\mathbf{w} +w_0)>1\\  -\lambda y^{(i)} (\mathbf{x}^{(i)})^\top & \text{Otherwise}
\end{cases}  $$

$$\nabla_{w_0} \text{LOSS}(\mathbf{w},w_0) =\sum_{i=1}^N\begin{cases} 0 & \text{if } \mathbf{y}^{(i)}( \mathbf{x}^{(i)}\mathbf{w} +w_0)>1\\  -\lambda y^{(i)}  & \text{Otherwise}
\end{cases}  $$


```python
from sklearn.datasets import make_blobs, make_circles, make_moons
from sklearn.preprocessing import StandardScaler

#some toy data generated by scikit-learn's
X_train, y = make_blobs(n_samples=500, centers=2,
                        n_features=2, random_state=1)
# We'll also scale the data to be centered at the origin with a unit standard deviation
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train, y)
# we have to change the class labels to be +1 and -1 instead of 0 and 1.
y=where(y<=0, -1,1)

X_outlier = np.append(X_train_scaled, [0.1, 0.1])
X_outlier = X_outlier.reshape(X_train.shape[0]+1, X_train.shape[1])
y_outlier = np.append(y, 1)

scatter(X_outlier[:,0],X_outlier[:,1],c=y_outlier)
```

```python
def predict(x, w, w0):
	activation = w.dot(x)+w0

	return 1 if activation >= 0.0 else -1
```

```python
def LOSS_f(X,y,w,w0,lambda_param):
   N,d = X.shape
   LOSS = 1/2 * norm(w)**2
   for i in range(N):
     condition = 1 - y[i] * (w.dot(X[i,:])+w0)
     LOSS = LOSS + lambda_param * max(condition, 0)
   return LOSS
```

```python
def train_weights(X, y, learning_rate_span,lambda_param, n_epoch=1000):
  N,d = X.shape
  w   = ones(2)*(-1)
  w[1] = 2
  w0  = 2
  for epoch in range(n_epoch):
    learning_rate = learning_rate_span[epoch]
    arr = np.arange(N)
    random.shuffle(arr)
    for i in arr:
          # condition = y[i] * (w.dot(X[i,:])+w0) >= 1
          condition = y[i] * (w.dot(X[i,:])+w0) > 1
          if condition:
            w = w - learning_rate *  w

          else:
            w  = w - learning_rate * (w - lambda_param *y[i] * X[i,:] )
            w0 = w0 - learning_rate * ( -lambda_param * y[i] )
    #print(iter)
  return w, w0



```

```python
n_epoch = 500
n_span = arange(n_epoch)
l_rate_span = 1/((1+ n_span)**2)
# w, w0 = train_weights(X_outlier, y_outlier, l_rate_span, lambda_param=10000, n_epoch=500)
w, w0 = train_weights(X_outlier, y_outlier, l_rate_span, lambda_param=1, n_epoch=500)

print(w)
print(w0)
```
